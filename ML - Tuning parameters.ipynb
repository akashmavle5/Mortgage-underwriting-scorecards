### Tuning parameters

#### tuning max_depth

Start with a wider range

```python
auc_dtree_list = []
auc_dtree_train_list = []

max_depth = [4, 6, 8, 10, 12]

for i in max_depth:

    Dtree = DecisionTreeClassifier(criterion="entropy", 
                                   max_depth=i, random_state = 42)
    Dtree = Dtree.fit(X_train, y_train)
    print('--------max_depth is:',i)
    #log_prob_model = -(Dtree.predict_log_proba(X_test)[:, 1] )*10
    #lower_score, upper_score, final_score=probhist(log_prob_model, y_test)
    #KS_max = KS_max(lower_score, upper_score, final_score)
    auc_dtree = auc(Dtree, X_test, y_test)
    auc_dtree_train = auc(Dtree, X_train, y_train)
    #KS_max_list.append(KS_max)
    auc_dtree_list.append(auc_dtree) 
    auc_dtree_train_list.append(auc_dtree_train)
    print('When C is:',i,'testing set AUC is:',auc_dtree,'training set AUC is:',auc_dtree_train)

plt.plot(max_depth, auc_dtree_list, c = 'red', label = 'test set AUC')
plt.plot(max_depth, auc_dtree_train_list, c = 'blue', label = 'train set AUC') 
plt.xlabel('max_depth', fontsize =12)
plt.ylabel('AUC score',fontsize =12)
plt.legend()
plt.show()
```

### Tuning parameters

Due to the computation time, I will tune one or two parameters each time and fixed other parameters for that.

```python
# for train and test score visualization 
def plot_results(model, param = 'n_estimators', name = 'Num Trees'):
    param_name = 'param_%s' % param

    # Extract information from the cross validation model
    train_scores = model.cv_results_['mean_train_AUC']
    test_scores = model.cv_results_['mean_test_AUC']
    train_time = model.cv_results_['mean_fit_time']
    param_values = list(model.cv_results_[param_name])
    
    # Plot the scores over the parameter
    plt.subplots(1, 2, figsize=(10, 6))
    plt.subplot(121)
    plt.plot(param_values, train_scores, 'bo-', label = 'train')
    plt.plot(param_values, test_scores, 'go-', label = 'test')
    plt.legend()
    plt.xlabel(name)
    plt.ylabel('Mean AUC')
    plt.title('Score vs %s' % name)
    
    plt.subplot(122)
    plt.plot(param_values, train_time, 'ro-')
    plt.xlabel(name)
    plt.ylabel('Train Time (sec)')
    plt.title('Training Time vs %s' % name)
    
    
    plt.tight_layout(pad = 4)
```

### Tuning by tree-based and boosting based

#### Define tuning fuction

```python
def GS_tune(parameters_test, X_train, y_train, file_name):
    print("# Tuning hyper-parameters")
    print()
    scores = {'AUC': 'roc_auc'}
    gbrt = GridSearchCV(GradientBoostingClassifier(random_state=42), parameters_test, #change para_test
                        scoring = scores, 
                        refit='AUC', return_train_score=True, verbose = 10)
    gbrt.fit(X_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(gbrt.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = gbrt.cv_results_['mean_test_AUC']
    stds = gbrt.cv_results_['std_test_AUC']
    for mean, std, params in zip(means, stds, gbrt.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
                % (mean, std * 2, params))
    gbrt.cv_results_ = pd.DataFrame(gbrt.cv_results_)
    out = file_name + '.csv'
    gbrt.cv_results_.to_csv(out)  
    return gbrt.cv_results_ 
```

# Xgboost

### Tuning parameters

Approach for XGB Parameter Tuning

- 1.Choose a relatively high learning rate. Generally a learning rate of 0.1 works but somewhere between 0.05 to 0.3 should work for different problems. Determine the optimum number of trees for this learning rate. XGBoost has a very useful function called as cv which performs cross-validation at each boosting iteration and thus returns the optimum number of trees required.
- 2.Tune tree-specific parameters ( max_depth, min_child_weight, gamma, subsample, colsample_bytree) for decided learning rate and number of trees. Note that we can choose different parameters to define a tree and Ill take up an example here.
- 3.Tune regularization parameters (lambda, alpha) for xgboost which can help reduce model complexity and enhance performance.
- 4.Lower the learning rate and decide the optimal parameters .

```python
def modelfit(alg, dtrain, dtest, X_cols, useTrainCV=True, cv_folds=5, early_stopping_rounds=50):
    
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgtrain = xgb.DMatrix(dtrain[X_cols].values, label=dtrain['d90_flag'].values)
        cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=alg.get_params()['n_estimators'], nfold=cv_folds,
            metrics='auc', early_stopping_rounds=early_stopping_rounds)
        alg.set_params(n_estimators=cvresult.shape[0])
    
    #Fit the algorithm on the data
    alg.fit(dtrain[X_cols], dtrain['d90_flag'],eval_metric='auc')
        
    #Predict training set:
    dtrain_predictions = alg.predict(dtrain[X_cols])
    dtrain_predprob = alg.predict_proba(dtrain[X_cols])[:,1]
    dtest_predprob = alg.predict_proba(dtest[X_cols])[:,1]
        
    #Print model report:
    print("\nModel Report")
    print("Accuracy : %.4g" % metrics.accuracy_score(dtrain['d90_flag'].values, dtrain_predictions))
    print("AUC Score (Train): %f" % metrics.roc_auc_score(dtrain['d90_flag'], dtrain_predprob))
    print("AUC Score (Test): %f" % metrics.roc_auc_score(dtest['d90_flag'], dtest_predprob))
                    
    feat_imp = pd.Series(alg._Booster.get_fscore()).sort_values(ascending=False)
    feat_imp.plot(kind='barh', title='Feature Importances')
    plt.ylabel('Feature Importance Score')
    return 
```

```python
def xgb_tune(gsearch):
    print("Best parameters set found on development set:")
    print()
    print(gsearch.best_params_)
    print(gsearch.best_score_)
    print()
    print("Grid scores on development set:")
    print()
    means = gsearch.cv_results_['mean_test_score'] 
    stds = gsearch.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, gsearch.cv_results_['params']):
        print("test score %0.3f (+/-%0.03f) for %r"
                % (mean, std * 2, params))
    return
```

```python
# for train and test score visualization 
def plot_results_xgb(model, param = 'n_estimators', name = 'Num Trees', pltime = False):
    param_name = 'param_%s' % param

    # Extract information from the cross validation model
    train_scores = model.cv_results_['mean_train_score']
    test_scores = model.cv_results_['mean_test_score']
    train_time = model.cv_results_['mean_fit_time']
    param_values = list(model.cv_results_[param_name])
    
    # Plot the scores over the parameter
    plt.subplots(1, 2, figsize=(10, 6))
    plt.subplot(121)
    plt.plot(param_values, train_scores, 'bo-', label = 'train')
    plt.plot(param_values, test_scores, 'go-', label = 'test')
    plt.legend()
    plt.xlabel(name)
    plt.ylabel('Mean AUC')
    plt.title('Score vs %s' % name)
    
    if pltime == True:
        plt.subplot(122)
        plt.plot(param_values, train_time, 'ro-')
        plt.xlabel(name)
        plt.ylabel('Train Time (sec)')
        plt.title('Training Time vs %s' % name)

    
    plt.tight_layout(pad = 4)
```

#### Step 1: Fix learning rate and number of estimators for tuning tree-based parameters

In order to decide on boosting parameters, we need to set some initial values of other parameters. Lets take the following values:

- max_depth = 5 : This should be between 3-10. Ive started with 5 but you can choose a different number as well. 4-6 can be good starting points.
- min_child_weight = 1 : A smaller value is chosen because it is a highly imbalanced class problem and leaf nodes can have smaller size groups.
- gamma = 0 : A smaller value like 0.1-0.2 can also be chosen for starting. This will anyways be tuned later.
- subsample, colsample_bytree = 0.8 : This is a commonly used used start value. Typical values range between 0.5-0.9.
- max_delta_step = 1: Because of high class imbalance. Since we care about predicting the right probability, we won't re-balance the dataset but set parameter maxdeltastep to a finite number (say 1) to help convergence instand

Please note that all the above are just initial estimates and will be tuned later. Lets take the default learning rate of 0.1 here and check the optimum number of trees using cv function of xgboost. The function defined above will do it for us.
